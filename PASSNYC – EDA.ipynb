Нам точно не нужны первые три колонки! Также для удобства вынесем оценки в отдельную колонку.

```python
schools_clear = schools.iloc[:, 3:]
schools_clear = schools_clear.loc[:, :'Student Achievement Rating']
schools_clear.head(2)
```

```python
grades = schools.loc[:, 'Average ELA Proficiency':]
grades.head(2)
```

Гораздо лучше!

```python
has_na = []
for column in schools_clear.columns:
    has_na.append(schools_clear[column].isnull().values.any())
```

```python
has_na = np.array(has_na)
schools_clear.columns[has_na == True]
```

```python
to_mean = schools_clear.columns[has_na == True]
schools_clear.loc[:, to_mean]
to_mean_int = schools_clear.loc[:, ['Economic Need Index', 'School Income Estimate', 'Student Attendance Rate', 'Percent of Students Chronically Absent', 'Rigorous Instruction %', 'Collaborative Teachers %', 'Supportive Environment %', 'Effective School Leadership %', 'Strong Family-Community Ties %','Trust %']]
```

```python
to_mean_int
```

```python
inc = to_mean_int['School Income Estimate']

from re import sub
from decimal import Decimal

keep = []
for element in inc:
    if type(element) == str:
        value = Decimal(sub(r'[^\d.]', '', element[1:-1]))
        keep.append(value)
    else:
        keep.append(0)
```

```python
keep = np.array(keep)
keep = keep.astype(np.float)
keep_new = list()

for element in keep:
    if element != np.float(0.0):
        keep_new.append(element)
    else: 
        keep_new.append(np.nan)
keep_new = np.array(keep_new)
```

```python
to_mean_int['School Income Estimate'] = keep_new
to_mean_int
```

```python
keep_new
```

```python
for name in ['Student Attendance Rate', 'Percent of Students Chronically Absent', 'Rigorous Instruction %', 'Collaborative Teachers %', 'Supportive Environment %', 'Effective School Leadership %', 'Strong Family-Community Ties %','Trust %']:
    keeper = []
    for element in to_mean_int[name]:
        if type(element) == str:
            keeper.append(int(element[:-1])/100)
        else:
            keeper.append(np.nan)
    keeper = np.array(keeper)
    to_mean_int[name] = keeper
    to_mean_int
```

```python
for column in to_mean_int.columns:
    to_mean_int[column] = to_mean_int[column].fillna(np.mean(to_mean_int[column]))
```

```python
to_mean_int
```

```python
type(to_mean_int['Student Attendance Rate'][0])
```

```python
for i in ['Economic Need Index', 'School Income Estimate', 'Student Attendance Rate', 'Percent of Students Chronically Absent', 'Rigorous Instruction %', 'Collaborative Teachers %', 'Supportive Environment %', 'Effective School Leadership %', 'Strong Family-Community Ties %','Trust %']:
    schools_clear[i] = to_mean_int[i]
```
